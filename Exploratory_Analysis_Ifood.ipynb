```python
# Running the database via SQL and BigQuery.

%%bigquery --project ifood-349417 

SELECT * from  ifood-349417.ifood.ifood 
```

```python
# Top 30 IDs with the highest income.

# Maybe ID 9432 is an outlier.

%%bigquery --project ifood-349417 

SELECT 
  ID,
  Income,
  Education,
  Marital_Status,
  Year_Birth,
  Teenhome,
  AcceptedCmp1,
  AcceptedCmp2,
  AcceptedCmp3,
  AcceptedCmp4,
  AcceptedCmp5



FROM ifood-349417.ifood.ifood 
ORDER BY Income desc
LIMIT 30
```

```python
# Top 30 IDs that most accepted campaigns.

# It was noticeable that families with higher incomes and fewer teenagers convert more site access into purchase.
# Being the niche that most accept campaigns. 

%%bigquery --project ifood-349417 
 
SELECT 
  ID,
  Income,
  Education,
  Marital_Status,
  Year_Birth,
  Teenhome,
  (AcceptedCmp1 +	AcceptedCmp2 +	AcceptedCmp3	+ AcceptedCmp4 + AcceptedCmp5) as Total_Acceptdcmp,
  NumWebPurchases,
  NumStorePurchases,
  NumWebVisitsMonth,
  MntWines,
  MntFruits,
  MntMeatProducts,
  MntFishProducts,
  MntSweetProducts,
  MntGoldProds
  



FROM ifood-349417.ifood.ifood
ORDER BY Total_Acceptdcmp desc
LIMIT 30
```

```python
#Top 30 IDs that least accepted campaigns.

# It was noticeable that families with lower incomes and more teenagers have a good number of access to the site,
# but most of them do not convert it into a purchase. This is the niche that least accepts campaigns.

%%bigquery --project ifood-349417 
SELECT 
  ID,
  Income,
  Education,
  Marital_Status,
  Year_Birth,
  Teenhome,
  (AcceptedCmp1 +	AcceptedCmp2 +	AcceptedCmp3	+ AcceptedCmp4 + AcceptedCmp5) as Total_Acceptdcmp,
  NumWebPurchases,
  NumStorePurchases,
  NumWebVisitsMonth,
  MntWines,
  MntFruits,
  MntMeatProducts,
  MntFishProducts,
  MntSweetProducts,
  MntGoldProds

FROM ifood-349417.ifood.ifood
ORDER BY Total_Acceptdcmp asc
LIMIT 30



```
