```python
# Running the database via SQL and BigQuery.

%%bigquery --project ifood-349417 

SELECT * from  ifood-349417.ifood.ifood 
```

```python
# Top 30 IDs with the highest income.

%%bigquery --project ifood-349417 

SELECT 
  ID,
  Income,
  Education,
  Marital_Status,
  Year_Birth,
  Teenhome,
  AcceptedCmp1,
  AcceptedCmp2,
  AcceptedCmp3,
  AcceptedCmp4,
  AcceptedCmp5



FROM ifood-349417.ifood.ifood 
ORDER BY Income desc
LIMIT 30
```

```python
# Top 50 IDs that most accepted campaigns.

# Families with higher incomes and less teenagers accept more campaigns.

%%bigquery --project ifood-349417
 
SELECT 
  ID,
  Income,
  Education,
  Marital_Status,
  Year_Birth,
  Teenhome,
  (AcceptedCmp1 +	AcceptedCmp2 +	AcceptedCmp3	+ AcceptedCmp4 + AcceptedCmp5) as Total_Acceptdcmp



FROM ifood-349417.ifood.ifood
ORDER BY Total_Acceptdcmp desc
LIMIT 50 
```

```python
#Top 50 IDs that least accepted campaigns.

# Families with lower income and more teenagers accept less campaigns.

%%bigquery --project ifood-349417 
SELECT 
  ID,
  Income,
  Education,
  Marital_Status,
  Year_Birth,
  Teenhome,
  (AcceptedCmp1 +	AcceptedCmp2 +	AcceptedCmp3	+ AcceptedCmp4 + AcceptedCmp5) as Total_Acceptdcmp

FROM ifood-349417.ifood.ifood
ORDER BY Total_Acceptdcmp asc
LIMIT 50



```
